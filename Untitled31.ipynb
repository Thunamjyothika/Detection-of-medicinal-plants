```python
import os

# List contents of the dataset directory
folder_path = "/content/Medicinal plant dataset"
files = os.listdir(folder_path)

for file in files:
    print(file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import numpy as np
import os
```

```python

# Step 1: Data Preprocessing and Augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1.0/255)
```

```python

# Dataset Paths (Ensure they exist)
train_dir = "/content/Medicinal plant dataset"
test_dir = "/content/Medicinal plant dataset"

if not os.path.exists(train_dir) or not os.path.exists(test_dir):
    raise FileNotFoundError("Dataset directory not found. Check the paths.")

train_set = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode="categorical"
)

test_set = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode="categorical"
)
```

```python

# Step 2: Build CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(40, activation='softmax')  # Adjusted for 40 classes
])
```

```python

# Step 3: Compile Model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python

# Step 4: Train Model
history = model.fit(
    train_set,
    epochs=50,  # Increased epochs
    validation_data=test_set,
    steps_per_epoch=int(len(train_set) / 255),
    validation_steps=int(len(test_set) / 255)
)
```

```python

# Step 5: Evaluate Model
loss, accuracy = model.evaluate(test_set)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```

```python
# Step 6: Save Model
model.save("medicinal_leaves_classifier.h5")
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt

def classify_image(image_path, model, train_set, confidence_threshold=0.1):
    if not os.path.exists(image_path):
        print(f"Error: Image file '{image_path}' not found.")
        return

    # Load and preprocess the image
    img = load_img(image_path, target_size=(128, 128))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0

    # Get predictions
    prediction = model.predict(img_array)
    predicted_class_index = np.argmax(prediction, axis=1)[0]
    confidence = np.max(prediction)

    # Ensure correct mapping from index to class
    class_labels = {v: k for k, v in train_set.class_indices.items()}
    predicted_class = class_labels.get(predicted_class_index, "Unknown")

    # Display the image
    plt.imshow(img)
    plt.title(f"Predicted Class: {predicted_class} (Confidence: {confidence:.2f})")
    plt.axis('off')
    plt.show()

    # Print classification result
    if confidence < confidence_threshold:
        print("Not a medicinal plant")
    else:
        print(f"Predicted Class: {predicted_class} (Confidence: {confidence:.2f})")

# Example usage (replace with your actual model, train_set, and image path)
# model = load_model('your_model.h5')
# train_set = ... (your ImageDataGenerator object)
# classify_image('/content/test1.jpg', model, train_set)
```
